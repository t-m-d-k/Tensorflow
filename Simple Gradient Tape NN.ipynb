```python
import tensorflow as tf
```

```python
w = tf.Variable(tf.random.normal((3, 1)), name='w')
b = tf.Variable(tf.zeros(3, dtype=tf.float32), name='b')
x = tf.Variable([[1., 2., 3.]])
y = tf.constant([2., 4., 6.])
```

```python
x.shape
```

```python
class MyModel(tf.Module):
  def __init__(self, **kwargs):
    super().__init__(**kwargs)
    # Initialize the weights to `5.0` and the bias to `0.0`
    # In practice, these should be randomly initialized
    g = tf.random.Generator.from_seed(1)
    self.w = tf.Variable(g.normal(shape=[3,1]))
    
    #self.w = tf.Variable(tf.random.normal((1)), name='w')
    self.b = tf.Variable(tf.zeros(3, dtype=tf.float32), name='b')

  def __call__(self, x):
    return ((x @ self.w)  + self.b)

model = MyModel()
```

```python
# This computes a single loss value for an entire batch
def loss(target_y, predicted_y):
  return tf.reduce_mean(tf.square(target_y - predicted_y))
```

```python
# Given a callable model, inputs, outputs, and a learning rate...
def train(model, x, y, learning_rate):

  with tf.GradientTape() as t:
    # Trainable variables are automatically tracked by GradientTape
    current_loss = loss(y, model(x))
    print("y", y.numpy())
    print("y_pred", model(x).numpy())
  # Use GradientTape to calculate the gradients with respect to W and b
  dw, db = t.gradient(current_loss, [model.w, model.b])

  # Subtract the gradient scaled by the learning rate
  model.w.assign_sub(learning_rate * dw)
  model.b.assign_sub(learning_rate * db)

  
```

```python
epochs = range(500)
```

```python
def training_loop(model, x, y):

  for epoch in epochs:
    # Update the model with the single giant batch
    train(model, x, y, learning_rate=0.01)

    # Track this before I update
#     weights.append(model.w.numpy())
#     biases.append(model.b.numpy())
    current_loss = loss(y, model(x))

    print(f"Epoch {epoch:2d}:")
    print("loss",current_loss)
  
  return model.w , model.b
    
    #print("    ", report(model, current_loss))
```

```python
w , b = training_loop(model, x, y)
```
